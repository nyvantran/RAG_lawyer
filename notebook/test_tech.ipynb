```python
from sentence_transformers import CrossEncoder
```

```python
model_name = 'keepitreal/vietnamese-sbert'  # Hoặc 'BAAI/bge-reranker-base'
print(f"Đang tải mô hình Re-ranker: {model_name}...")
reranker = CrossEncoder(model_name)
```

```python
query = "Thành phố nào là thủ đô của Việt Nam?".upper()

documents = [
    "Thành phố Hồ Chí Minh là trung tâm kinh tế lớn nhất của Việt Nam.",
    "Hà Nội là thủ đô của nước Cộng hòa Xã hội Chủ nghĩa Việt Nam, có lịch sử hơn ngàn năm văn hiến.",
    "Đà Nẵng được mệnh danh là thành phố đáng sống nhất Việt Nam với nhiều cây cầu nổi tiếng.",
    "Phở là một món ăn truyền thống nổi tiếng của người Việt Nam, đặc biệt là ở Hà Nội.",
    "Thủ đô của một quốc gia thường là nơi đặt các cơ quan chính phủ trung ương."
]
```

```python
print(f"\n[Câu hỏi]: {query}")
print("-" * 50)
print("Đang tiến hành chấm điểm và tái xếp hạng...")
scores = reranker.rank(
    query=query,
    documents=documents,
)
for score in scores:
    print(f"{score["score"]}: {documents[score["corpus_id"]]}")
print(documents)
```
